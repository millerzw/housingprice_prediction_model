```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("data\housing.csv")
```

# Pre-Processing

### First we must ensure that the data is not missing any values

```python
data.info()
```

Given we are missing some values in total_bedrooms, we must fix this before continuing, We will drop the data points that contain the null values

```python
data.dropna(inplace=True)
```

```python
data.info()
```

### Split the Data into Train and Test Versions

```python
solve_for = 'median_house_value'

X = data.drop([solve_for], axis=1)
y = data[solve_for]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2)
```

```python
train_data = X_train.join(y_train)
train_data
```

```python
train_data.hist(figsize= (15,10))
```

## HeatMap for Identifying Trends Between Data Catregories

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(numeric_only = True), annot=True, cmap="mako")
```

# Normalize The Data
The data is right skewed for many of the categories, and so we will take the logarithm of these factors to push our values into a gaussian distribution

```python
data['total_rooms'] = np.log(data['total_rooms']+1)
data['total_bedrooms'] = np.log(data['total_bedrooms']+1)
data['population'] = np.log(data['population']+1)
data['households'] = np.log(data['households']+1)

```

```python
data.hist(figsize= (15,10))
```

### One-hot Encode the Non-Numeric Features
We must one hot encode the non-integer features in order to include them in our prediction model without skewing their importance

```python
data.ocean_proximity.value_counts()
```

```python
data = train_data.join(pd.get_dummies(data.ocean_proximity))
data.drop(['ocean_proximity'], axis=1, inplace=True)
```

```python
data
```

### Show The Correlation of the data

```python
plt.figure(figsize=(15,10))
sns.heatmap(data.corr(numeric_only = True), annot=True, cmap="mako")
```

## Visualizing The Pricing
Notice that we now see a positive correlation between price and distance to the ocean, but notice the lack of correlation of 'island' homes

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='latitude',y='longitude', data=data, hue='median_house_value', palette='coolwarm')
```

#### * Plotting the coordinates of each housing tract along with their median house value shows the relationship between distance to the coast and price more clearly *

# Train Our Prediction Model

## Linear Regresssion

```python
from sklearn.linear_model import LinearRegression

solve_for = 'median_house_value'

X = data.drop([solve_for], axis=1)
y = data[solve_for]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2)

regr = LinearRegression()

regr.fit(X_train, y_train)

train_data = X_train.join(y_train)
test_data = X_test.join(y_test)
```

### Get the validation of the Linear Regression Model

```python
regr.score(X_test, y_test)
```
