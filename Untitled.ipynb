# Normalize The Data
> The data is right skewed for many of the categories, and so we will take the logarithm of these factors to push our values into a gaussian distribution

```python
data['total_rooms'] = np.log(data['total_rooms']+1)
data['total_bedrooms'] = np.log(data['total_bedrooms']+1)
data['population'] = np.log(data['population']+1)
data['households'] = np.log(data['households']+1)

```

```python
data.hist(figsize= (15,10))
```

### One-hot Encode the Non-Numeric Features
> We must one hot encode the non-integer features in order to include them in our prediction model without skewing their importance

```python
data.ocean_proximity.value_counts()
```

```python
data = data.join(pd.get_dummies(data.ocean_proximity))
data.drop(['ocean_proximity'], axis=1, inplace=True)
```

# Train Our Prediction Model

## Linear Regresssion
> We will create a model based on linear regression aka a best fit line upon the multi-dimensional data

### Training The Model

```python
from sklearn.linear_model import LinearRegression

solve_for = 'median_house_value'

X = data.drop([solve_for], axis=1)
y = data[solve_for]

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2)

regr = LinearRegression()

regr.fit(X_train, y_train)

train_data = X_train.join(y_train)
test_data = X_test.join(y_test)
```

### Get the validation of the Linear Regression Model

```python
regr.score(X_test, y_test)
```

>Depending how the data is split, we get about **65%-67%** accuracy, this is fairly poor

## Decision Tree Classifier (Random Forest Regression)
> We will create a model based on decision trees to classify the data

### Training The Model

```python
from sklearn.ensemble import RandomForestRegressor

forestReg = RandomForestRegressor()

forestReg.fit(X_train, y_train)
```

### Get the Validation of the Random Forest Regression Model

```python
forestReg.score(X_test, y_test)
```

> Depending how the data is split, we get about **80%-84%** accuracy, this can be improved with hyperparameter tuning, and is much better than a standard regression model already

### HyperParameter Tuning using Cross-Validation

```python
from sklearn.model_selection import GridSearchCV

forestReg = RandomForestRegressor()
```

#### Creating our Parameter Grid to Test From

```python
# Number of trees to use
n_estimators = [int(x) for x in np.linspace(start = 50, stop = 500, num = 5)]

# Number of features to consider at every split
max_features = [int(x) for x in np.linspace(start = 2, stop = 10, num = 5)]

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 100, num = 5)]
max_depth.append(None)

# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]

# Create the random grid
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

### Applying the Model

```python
# Using a GridSearch on our Forest Tree Regression with our new paramter grid and 5 fold cross validation
grid_search = GridSearchCV(forestReg, param_grid, cv=5, scoring = "neg_mean_squared_error", return_train_score=True)

grid_search.fit(X_train, y_train)
```
